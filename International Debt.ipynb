```python
#Total debt by all countries in the tables
%sql SELECT SUM(debt) as sum_debt FROM international_debt;
```

```python
#Top 10 countries have the most debt in the tables
%sql SELECT country_name,SUM(debt) as sum_debt FROM international_debt GROUP BY 1 ORDER BY 2 DESC LIMIT 10;
```

```python
#Top 10 countries have the least debt in the tables
%sql SELECT country_name,SUM(debt) as sum_debt FROM international_debt GROUP BY 1 ORDER BY 2 ASC LIMIT 10;
```

```python
# Top 10 Total debt classified by indicators
%sql SELECT indicator_name, indicator_code, SUM(debt) as sum_debt from international_debt GROUP BY 1,2 ORDER BY 3 DESC LIMIT 10;
```

```python
# Top 10 AVG debt classified by indicators
%sql SELECT indicator_name, indicator_code, AVG(debt) as avg_debt from international_debt GROUP BY 1,2 ORDER BY 3 DESC LIMIT 10;
```
